```python
with open('/content/drive/MyDrive/HashCode/2020/Hash Code 2020 - Online Qualification Round/c_incunabula.txt') as f:
    lines = f.readlines()
```

```python
config = {'TotalBook': 0, 'Lib':0 ,  'Days': 0}
key=lines[0].split()
config["TotalBook"]=int(key[0])
config["Lib"]=int(key[1])
config["Days"]=int(key[2])
```

```python
class Book:
  switch=1
  def __init__(self, x): 
        self.Score = x 

Book_List=[]
key=lines[1].split()
for i in key:
  Book_List.append(Book(int(i)))
```

```python
class Library:
  signup_end=0
  Odered_book=[]
  switch=0
  def __init__(self, x,y,z,k): 
        self.Total_Book = x 
        self.Sign_process = y 
        self.quantity = z
        self.ListBook = k

Lib_Detail=[]
for i in range(1,config["Lib"]+1):
  j=i*2
  key=lines[j].split()
  templist=lines[j+1].split()
  for k in range(0,len(templist)):
    templist[k]=int(templist[k])

  Lib_Detail.append(Library(int(key[0]),int(key[1]),int(key[2]),templist)) 
```

```python
eff_lib=[]
for i in range(0,len(Lib_Detail)):
  eff_lib.append(Lib_Detail[i].Sign_process)
temp=eff_lib.copy()
eff_lib.sort()

Preffer_Lib=[]
for i in eff_lib:
  Preffer_Lib.append(temp.index(i))
  temp[temp.index(i)]=-1
```

```python
time=0
for i in range(0,len(Preffer_Lib)):
  Lib_Detail[Preffer_Lib[i]].signup_end= time +Lib_Detail[Preffer_Lib[i]].Sign_process
  time=Lib_Detail[Preffer_Lib[i]].signup_end
```

```python
temp=[]
for i in range(0,len(Book_List)):
  temp.append(Book_List[i].Score)
temp2=temp.copy()
temp.sort()

Preffer_book=[]
for i in temp:
  Preffer_book.append(temp2.index(i))
  temp2[temp2.index(i)]=-1
Preffer_book=Preffer_book[::-1]
```

```python
def time_req(x):
  y=Lib_Detail[x].signup_end+int( len(Lib_Detail[x].Odered_book) / Lib_Detail[x].quantity )+1
  return(y)
```

```python
for book in Preffer_book:
  temp_lib=[]
  temp_time=[]

  for lib in Preffer_Lib:
    if book in Lib_Detail[lib].ListBook:
      temp_time.append(time_req(lib))
      temp_lib.append(lib)

  temp2_time=temp_time.copy()
  temp_time.sort()
  best_lib=temp_lib[temp2_time.index(temp_time[0])]
  cpy=Lib_Detail[best_lib].Odered_book.copy()
  cpy.append(book)
  Lib_Detail[best_lib].Odered_book=cpy
```
